# CyTOF and Statistics
One of the things that always troubled me about Flow Cytometry is the seemingly arbitrary coordinate transforms applied to their data: the "symmetrical log", "logical", "arcsinh with cofactor five".  It wasn't that they used these transforms when plotting data, it was that they first transformed their data and then performed standard statistical operations: density estimation, histograms, etc.  If you're not careful, this can introduce spurious artifacts.  However, if you don't, your figures don't reveal anything.

```python
import flowml as fml
import matplotlib.pyplot as plt
fd = fml.FlowData('test.fcs')
fig, ax = plt.subplots(1,2, figsize = (13,4))
fd.hist('Beads','Ir191', axes = ax[0])
fd.hist('Beads','Ir191', axes = ax[1], transform = 'linear')
ax[0].set_title('Arcsinh Transformation')
ax[1].set_title('Linear Transformation');
```

By using an 'arcsinh' transform on the left, the data is transformed by the relation
\begin{equation}
    \DeclareMathOperator{\arcsinh}{arcsinh}
    \hat{x} = \arcsinh (5 x)
\end{equation}
and then binned into the histogram as seen on the left.  On the right, no such transformation was applied.  The main consequence is the doublet population is not obvious.  Instead, putative doublets appear in a diffuse clump at the bottom of the plot.  Neither plot shows the negative population well.

## Why Transforming and then Binning is a Bad Idea
The problem with transforming the data, binning in that transformed space and then plotting again is a bad idea is that this can give a false sense of a dense population.

## What is Really Going On
I've asked around and no one has worked out what the statistics are of markers on the surface.  No one knows what the distribution of the count of surface markers looks like on these cells.  We also have to content with the fact that antibody binding is often not controlled for.  It can be with beads with antigens on the surface, but this is rarely done.  

First, cells typically have large numbers of antibodies.  For the CyTOF to detect even one count, there must be at least 1000 tagged antibodies bound.  So when we see counts in the tens and thousands, we know there are lots of surface markers.  The law of large numbers then suggests that, under reasonable assumptions, the number of markers should be normally distributed.  But how many do we actually see?  Antibody binding is reversable, so after washing, these may start to unbind.  This process is a Poisson process, but as there are so many sites, we can essentially consider it normally distributed.

What we see then is essentially the combination of two Gaussian distributions, one whose variance is porportional to its mean, and the other which is intrinsic and varying between cell types and expected counts.
